---
## 1. 문자열 → 날짜 변환


````python
import pandas as pd
````

````python
df = pd.read_csv('./data/all_data.csv', index_col= 0)
````

````python
df.head()
````

````python
df['주문일자'] = pd.to_datetime( df['주문일자'] )
````

````python
df.info()
````

---
## 2. `dt` accessor - 날짜 요소 추출

`.dt.year`, `.dt.month`, `.dt.day`, `.dt.dayofweek` 등 다양한 요소를 꺼낼 수 있습니다.


````python
df['년도']  =  df['주문일자'].dt.year
````

````python
df.head(1)
````

````python
df['월'] = df['주문일자'].dt.month
````

````python
df['날짜'] = df['주문일자'].dt.day
````

````python
# 0 : 월요일,  1 : 화요일, ...  6 : 일요일
df['요일'] = df['주문일자'].dt.dayofweek
````

---
## 3. 날짜 필터링


````python
# 1분기 주문 데이터를 가져오세요. 
df[ (df['월'] >= 1 ) & (df['월'] <= 3) ]
````

````python
df[ df['요일'] == 4 ]['매출'].sum()
````

---
## 4. 월별/요일별 집계


````python
# 각 월별로 매출총합 구해줘
df.groupby('월')['매출'].sum()
````

````python
# 요일별로 주문건수를 확인하고 싶다.  어느요일에 주문이 많나????
df.groupby('요일')['주문번호'].count()
````

````python
df['요일'].value_counts()
````

---
## 5. `resample` - 시계열 재샘플링

날짜를 인덱스로 설정하면 `resample`로 월별/주별 집계를 더 편하게 할 수 있습니다.


````python
df.head(2)
````

````python
# 1. 시간컬럼을 인덱스로 만든다.
df = df.set_index('주문일자')
````

````python
df.head(3)
````

````python
# 2. 시간컬럼을 인덱스로 만들면, 
#    일단위, 주단위, 월말단위, 분기말 단위, 연말 단위, 초단위, 분단위, 시간단위로 데이터 분석가능
#    옵션: "D"(일), "W"(주), "ME"(월말), "QE"(분기말), "YE"(연말)
````

````python
# 월말단위로 매출 총합을 구하라.
df['매출'].resample('ME').sum()
````

````python
# 주 단위로 매출 총합을 구하라
df['매출'].resample('W').sum()
````
