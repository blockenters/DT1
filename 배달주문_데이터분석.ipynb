## 프로젝트 개요

### 배경
국내 배달 시장은 최근 몇 년간 폭발적으로 성장하며, 배달 플랫폼의 데이터 활용 역량이 핵심 경쟁력이 되었습니다.  
본 프로젝트에서는 가상의 배달 플랫폼 데이터를 활용하여 **매출, 고객, 배달 성과, 날씨 영향** 등을 종합적으로 분석합니다.

### 목표
1. 데이터 전처리를 통해 **분석 가능한 품질**의 데이터를 확보한다.
2. 다양한 관점(매출, 시간, 메뉴, 배달, 고객, 날씨)에서 **인사이트**를 도출한다.
3. 도출된 인사이트를 바탕으로 **비즈니스 제안**을 제시한다.

### 데이터 설명

| 파일 | 설명 | 주요 컬럼 | 행 수 |
|------|------|-----------|-------|
| `stores.csv` | 가게 정보 | 가게ID, 가게명, 업종, 지역, 평균배달시간, 가게평점 | 100 |
| `customers.csv` | 고객 정보 | 고객ID, 성별, 연령, 지역, 가입일, 등급 | 500 |
| `orders.csv` | 주문 정보 | 주문ID, 주문일시, 고객ID, 가게ID, 메뉴, 주문금액, 수량, 배달소요시간, 결제방법, 쿠폰사용여부, 평점, 주문상태 | 5,000 |

- **비수기 매출 보완**: 주문이 적은 시간대(아침, 오후)에 할인 쿠폰을 발행하여 수요를 분산시킵니다.

### 2. 배달 서비스 개선
- **배달시간 단축**: 배달 소요시간과 평점이 음의 상관관계를 보이므로, 배달시간 단축이 고객 만족도 향상에 직접적으로 기여합니다.
- **배달 지연 지역 집중 관리**: 평균 배달시간이 높은 지역에 라이더 배치를 늘리거나, 가게-고객 간 거리 최적화를 진행합니다.
- **취소율 낮추기**: 취소율이 높은 업종과 시간대를 분석하여, 예상 배달시간을 정확하게 안내하고 대기 시간 알림을 강화합니다.

### 3. 고객 관리 전략
- **VIP 고객 관리**: 상위 등급 고객의 평균 주문금액이 높으므로, 맞춤형 혜택(무료 배달, 전용 쿠폰)으로 이탈을 방지합니다.
- **재주문률 향상**: 1회 주문 고객에게 재주문 유도 쿠폰을 발송하여, 충성 고객으로 전환합니다.
- **쿠폰 전략 최적화**: 쿠폰 사용 시 주문금액 변화를 모니터링하여, ROI가 높은 쿠폰 타입을 발굴합니다.

### 4. 날씨 대응 전략
- **우천 시 프로모션**: 비/눈 오는 날 주문이 늘어나는 업종에 대해 맞춤 프로모션을 자동 발동합니다.
- **기상 기반 수요 예측**: 기온, 강수량 데이터를 활용하여 일별 주문량을 예측하고, 라이더 수급을 사전에 조절합니다.
- **계절별 메뉴 추천**: 기온에 따라 인기 메뉴가 달라지므로, 계절별 추천 메뉴 큐레이션을 도입합니다.
