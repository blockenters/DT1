## 3. 패키지 설치와 관리 (pip / conda)

외부 패키지는 설치해야 사용할 수 있습니다.

### conda (미니콘다) - 우리가 사용할 방식
```bash
# 패키지 설치
conda install numpy pandas matplotlib seaborn scikit-learn

# 가상환경 생성 (프로젝트별 독립 환경)
conda create -n data_analysis python=3.10
conda activate data_analysis

# 설치된 패키지 확인
conda list
```

### pip (파이썬 기본 패키지 관리자)
```bash
# 패키지 설치
pip install numpy pandas

# 설치된 패키지 확인
pip list
```

> **실무 팁**: conda와 pip를 섞어 쓰면 충돌이 날 수 있습니다.  
> 가능하면 **conda를 우선** 사용하고, conda에 없는 패키지만 pip로 설치하세요.
