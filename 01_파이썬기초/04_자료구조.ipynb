## 왜 배워야 하나요?

데이터분석은 결국 **데이터를 담고, 꺼내고, 변환하는** 작업입니다.  
- **리스트**: 여러 데이터를 순서대로 저장 (매출 목록, 고객 목록 등)
- **딕셔너리**: 키-값 쌍으로 저장 (고객 정보, 설정값, JSON 데이터 등)
- **튜플**: 변경 불가능한 묶음 (좌표, 함수 반환값 등)
- **셋**: 중복 제거, 집합 연산 (고유값 추출, 교집합 분석 등)

Pandas의 DataFrame도 내부적으로 이 자료구조들을 기반으로 동작합니다.

## 5. 자료구조 선택 가이드

| 상황 | 자료구조 | 이유 |
|------|----------|------|
| 순서가 있는 데이터 모음 | 리스트 | 인덱싱, 슬라이싱 가능 |
| 키로 값을 찾아야 할 때 | 딕셔너리 | 빠른 검색, 직관적 |
| 변하면 안 되는 데이터 | 튜플 | 불변성 보장 |
