# 02. 조건문과 반복문

## 학습 목표
- if/elif/else 조건문으로 데이터를 분류할 수 있다
- for/while 반복문으로 데이터를 순회하며 처리할 수 있다
- 리스트 컴프리헨션으로 간결하게 데이터를 변환할 수 있다

---

## 왜 배워야 하나요?

데이터분석에서 **조건에 따라 데이터를 분류**하고, **대량의 데이터를 반복 처리**하는 작업은 기본 중의 기본입니다.  
- 조건문: 고객 등급 분류, 이상치 판별, 카테고리 매핑 등
- 반복문: 여러 파일 일괄 처리, 데이터 순회, 배치 분석 등

Pandas를 쓰더라도, 복잡한 조건 로직이나 파일 단위 처리에서는 반드시 조건문과 반복문이 필요합니다.
