## 2. 이상치 탐지 (IQR 방식)

**IQR(Interquartile Range) 방식**이 가장 일반적입니다.
- Q1 = 25% 분위수, Q3 = 75% 분위수
- IQR = Q3 - Q1
- 정상 범위: `[Q1 - 1.5*IQR, Q3 + 1.5*IQR]`
- 이 범위 밖이면 이상치로 간주
