## 왜 배워야 하나요?

실무에서 "이 데이터 분석해주세요"라고 받았을 때,

### EDA 프로세스

```
1. 데이터 로딩 & 구조 파악
2. 데이터 품질 확인 (결측, 이상치, 타입)
3. 전처리
4. 단변량 분석 (변수 하나씩 분포 확인)

6. 시각화 & 인사이트 도출
```
