## 왜 배워야 하나요?

데이터분석에서 다루는 데이터는 대부분 **숫자로 이루어진 표(테이블)** 형태입니다.  
NumPy는 이런 숫자 데이터를 **빠르고 효율적으로** 처리하기 위한 라이브러리입니다.  

Python의 리스트로도 숫자 계산을 할 수 있지만, 데이터가 수만~수백만 건이 되면 속도 차이가 엄청납니다.  
NumPy는 내부적으로 C언어로 구현되어 있어서 **리스트보다 수십~수백 배 빠릅니다.**  

또한 Pandas, Scikit-learn 등 데이터분석/머신러닝 라이브러리가 모두 NumPy 위에 만들어져 있기 때문에,  
NumPy를 이해하면 이후 모든 학습이 수월해집니다.

## 6. 통계 함수

데이터분석의 시작은 **기초 통계량 확인**입니다.  
NumPy는 평균, 합계, 최대/최소, 표준편차 등 핵심 통계 함수를 제공합니다.

| 함수 | 설명 | 실무 활용 |
|------|------|----------|
| `np.mean()` | 평균 | 평균 매출, 평균 체류시간 |
| `np.sum()` | 합계 | 총 매출, 총 주문건수 |
| `np.max()`, `np.min()` | 최대/최소 | 최고 매출일, 최저 재고 |
| `np.std()` | 표준편차 | 매출 변동성, 품질 편차 |
| `np.median()` | 중앙값 | 중위 소득, 중위 가격 |
| `np.argmax()`, `np.argmin()` | 최대/최소 인덱스 | 최고 매출 월 찾기 |

monthly_sales = np.array([2800, 3200, 3500, 3100, 3800, 4200, 3900, 4500, 4100, 3700, 3300, 5100])  # 만원
