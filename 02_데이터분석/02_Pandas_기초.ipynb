## 2. DataFrame

DataFrame은 **행과 열로 이루어진 2차원 테이블**입니다.  
엑셀 시트, SQL 테이블, CSV 파일과 같은 구조입니다.

Pandas에서 **가장 핵심적인 자료구조**이며, 데이터분석의 거의 모든 작업이 DataFrame 위에서 이루어집니다.

| 구성요소 | 설명 |
|---------|------|
| index | 행 이름 (0, 1, 2... 또는 라벨) |
| columns | 열 이름 (변수명) |

## 3. 데이터 선택 (열/행 접근)

DataFrame에서 원하는 데이터를 꺼내는 방법입니다.  
크게 **열 선택**과 **행 선택** 두 가지로 나뉩니다.

| 방법 | 문법 | 설명 |
|------|------|------|
| 열 선택 | `df["컬럼명"]` | 한 열 (Series 반환) |
| 열 선택 | `df[["컬럼1", "컬럼2"]]` | 여러 열 (DataFrame 반환) |
| 행 선택 | `df.loc[라벨]` | 인덱스 라벨로 행 선택 |
| 행 선택 | `df.iloc[번호]` | 위치 번호로 행 선택 |

```python
data = {
    "이름": ["김민수", "이영희", "박철수", "최지은", "정현우"],
    "부서": ["영업팀", "개발팀", "영업팀", "마케팅팀", "개발팀"],
    "연봉": [4200, 5100, 3800, 4500, 4800],
    "입사연도": [2019, 2017, 2021, 2018, 2020],
    "평가점수": [85, 92, 78, 88, 95]
}
df = pd.DataFrame(data)
```

````python
# 열 선택: df["컬럼명"] → Series 반환
# 언제 쓰나: 특정 변수(열)만 꺼내서 분석할 때 (예: 연봉 분포 확인)

# 아래에 코드를 작성하세요
````
