## 왜 배워야 하나요?

NumPy는 숫자 배열에 특화되어 있지만, 실무 데이터는 **이름(라벨)** 이 붙어 있어야 편리합니다.
"강남점 매출", "2024년 1월 매출"처럼 라벨로 접근할 수 있어야 분석이 직관적입니다.

**Series**는 **인덱스(이름)가 붙은 1차원 배열**입니다. 엑셀의 한 컬럼을 생각하면 됩니다.

> **실무 팁**: Series를 직접 만드는 경우는 드물지만, DataFrame에서 컬럼 하나를 꺼내면 Series가 됩니다.
