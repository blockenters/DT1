SQL의 `JOIN`과 동일한 개념입니다.

예시:
- 주문 테이블의 `고객ID` → 고객 테이블에서 고객 이름, 지역 가져오기

| merge 방식 | SQL 대응 | 설명 |
|-----------|---------|------|
| `how='inner'` | INNER JOIN | 양쪽 모두 있는 것만 |

| `how='right'` | RIGHT JOIN | 오른쪽 기준 |
| `how='outer'` | FULL OUTER JOIN | 양쪽 모두 포함 |
