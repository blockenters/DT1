결측치(Missing Value)는 **값이 없는 데이터**입니다.  
Pandas에서는 `NaN`(Not a Number)으로 표현됩니다.

### 결측치 처리 전략

| 방법 | 함수 | 언제 사용 |
|------|------|----------|
| 확인 | `isna()`, `isnull()` | 결측치 위치/개수 파악 |
