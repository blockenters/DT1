## 왜 배워야 하나요?

**DataFrame**은 행과 열로 이루어진 2차원 테이블입니다. 엑셀 시트, SQL 테이블과 같은 구조입니다.
Pandas에서 **가장 핵심이 되는 자료구조**이며, 데이터분석의 거의 모든 작업이 DataFrame 위에서 이루어집니다.

| 구성요소 | 설명 |
|---------|------|
| index | 행 이름 (0, 1, 2... 또는 라벨) |
| columns | 열 이름 (변수명) |
| values | 실제 데이터 |
