## 왜 배워야 하나요?

숫자로만 보면 놓치는 것들이 **그래프로 보면 한눈에** 보입니다.  
"매출이 꾸준히 성장하고 있다", "20대 고객이 가장 많다", "이상치가 있다" 같은 패턴은  
표보다 그래프로 훨씬 빠르게 파악할 수 있습니다.

또한 분석 결과를 **상사나 팀원에게 전달**할 때, 그래프 없이 숫자만 나열하면  
아무도 관심을 갖지 않습니다. 시각화는 **커뮤니케이션 도구**입니다.

### 차트 선택 가이드

| 보고 싶은 것 | 적합한 차트 |
|-------------|----------|
| 시간에 따른 변화 | 선 그래프 |
| 카테고리별 비교 | 막대 그래프 |
| 데이터 분포 | 히스토그램, 박스플롯 |
| 두 변수의 관계 | 산점도 |
| 비율/구성 | 파이 차트 |
| 변수 간 상관관계 | 히트맵 |
