## 프로젝트 개요

### 배경
구독 기반 서비스 시장이 성장하면서, 신규 고객 확보보다 기존 고객 유지가 비용 효율적이라는 인식이 확산되고 있습니다. 고객 이탈률을 1%만 줄여도 수익에 큰 영향을 미치며, 이를 위해서는 고객을 체계적으로 세분화하고 이탈 가능성을 사전에 예측하는 것이 핵심입니다.

### 목표
1. **고객 세분화**: RFM 분석과 K-Means 클러스터링을 활용하여 고객군을 식별합니다.
2. **이탈 예측**: 머신러닝 모델을 통해 이탈 가능성이 높은 고객을 사전에 탐지합니다.
3. **비즈니스 전략**: 세분화 결과와 이탈 예측을 결합하여 맞춤형 리텐션 전략을 제안합니다.

### 6.4 전략 제안

#### 고객 등급별 리텐션 전략

| 등급 | 전략 | 구체적 액션 |
|------|------|-------------|
| **VIP 고객** | 프리미엄 유지 | 전담 매니저 배정, 독점 혜택 제공, 정기 만족도 조사 |
| **우수 고객** | 업셀링/크로스셀링 | 상위 요금제 업그레이드 유도, 번들 상품 추천 |
| **일반 고객** | 활성화 촉진 | 사용 패턴 기반 맞춤 콘텐츠 추천, 리워드 프로그램 |
| **이탈위험 고객** | 긴급 대응 | 할인 쿠폰 발송, 1:1 상담, 불만 사항 즉시 해결 |

#### 이탈 위험 등급별 대응 방안

- **고위험 (이탈확률 70% 이상)**: 즉시 전담 상담원 연결, 특별 할인 또는 무료 서비스 기간 제공
- **중위험 (이탈확률 40~70%)**: 맞춤형 프로모션 발송, 서비스 이용 가이드 제공
- **저위험 (이탈확률 40% 미만)**: 정기적 만족도 모니터링, 로열티 프로그램 운영

#### 핵심 인사이트

1. **로그인 빈도 감소**가 이탈의 가장 강력한 신호입니다. 로그인 횟수가 급감하는 고객을 조기에 감지하는 알림 시스템 구축이 필요합니다.
2. **불만 횟수 증가**는 이탈의 주요 선행 지표입니다. CS 응대 품질 향상과 불만 사항 신속 처리 프로세스가 중요합니다.
3. **거래 빈도(Frequency)**와 **최근성(Recency)**이 고객 가치와 이탈 예측 모두에 핵심적인 역할을 합니다.
4. 월 단위 모니터링을 통해 이탈 예측 모델을 주기적으로 업데이트하고, 위험 등급 변화를 추적하는 것을 권장합니다.
