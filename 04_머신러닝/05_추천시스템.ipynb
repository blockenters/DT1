### 개념 설명: 추천 시스템이란?

**추천 시스템 = "이 유저가 좋아할 만한 상품을 자동으로 찾아주는 것"**

우리가 매일 접하는 추천 시스템 예시:
- **넷플릭스**: "이 영화 어때요?"
- **쿠팡**: "이 상품과 비슷한 상품"
- **유튜브**: "추천 영상"
- **스포티파이**: "취향 맞춤 플레이리스트"

#### 추천 시스템의 3가지 방식 비교

| 방식 | 원리 | 장점 | 단점 |
|------|------|------|------|
| **인기도 기반** | 많이 팔리고 평점 높은 거 추천 | 심플하고 콜드스타트 문제 없음 | 개인화가 안 됨 (모두에게 같은 추천) |
| **유저 기반 협업필터링** | 나와 취향 비슷한 사람이 좋아한 것 추천 | 개인화됨 | 신규 유저 문제 (데이터가 없으면 추천 불가) |
| **아이템 기반 협업필터링** | 이 상품과 비슷한 상품 추천 | 안정적이고 실무에서 많이 사용 | 신규 상품 문제 (데이터가 없으면 추천 불가) |

> **콜드스타트(Cold Start) 문제란?**
> 새로운 유저나 새로운 상품이 들어왔을 때, 아직 평점/구매 데이터가 없어서 추천을 못 하는 문제입니다.
> 인기도 기반은 개인 데이터가 필요 없으므로 이 문제가 없습니다!

### 개념 설명: 유저 기반 협업 필터링 원리

**핵심 아이디어: "나와 비슷한 취향의 사람이 좋아한 것 = 나도 좋아할 확률 높음"**

예를 들어, 내가 영화 A, B, C를 좋아하는데, 철수도 A, B, C를 좋아한다면?
→ 철수가 좋아하는 영화 D를 나에게 추천하면 좋아할 가능성이 높습니다!

#### 코사인 유사도(Cosine Similarity)란?
- 두 유저의 **평점 패턴이 얼마나 비슷한지** 측정하는 방법입니다
- **0 ~ 1** 사이의 값으로, **1에 가까울수록 취향이 완전히 같다**는 뜻입니다
- 예: 유사도 0.95 = "거의 같은 취향", 유사도 0.1 = "취향이 매우 다름"

#### 유저 기반 협업 필터링 과정

1. **유저-상품 평점 매트릭스 만들기** (행: 유저, 열: 상품, 값: 평점)
2. **유저 간 코사인 유사도 계산** (모든 유저 쌍의 유사도를 구함)
3. **나와 가장 유사한 유저 Top N 찾기** (유사도가 높은 순서로)
4. **그 유저들이 높게 평가했지만 내가 안 본 상품 = 추천!**

### 개념 설명: 아이템 기반 협업 필터링 원리

**핵심 아이디어: "이 상품을 좋아한 사람들이 저 상품도 좋아함 = 비슷한 상품"**

예를 들어, 아이폰 케이스를 산 사람들이 아이폰 충전기도 많이 샀다면?
→ 아이폰 케이스를 보고 있는 유저에게 "아이폰 충전기도 어때요?" 라고 추천합니다.

#### 유저 기반 vs 아이템 기반 비교

| 구분 | 유저 기반 | 아이템 기반 |
|------|----------|-----------|
| 유사도 계산 대상 | 유저 간 유사도 | 상품 간 유사도 |
| 안정성 | 유저가 변하면 추천도 바뀜 (불안정) | 상품 관계는 비교적 안정적 |
| 실무 활용 | 덜 사용됨 | **실무에서 더 많이 사용** |
| 대표 서비스 | - | 아마존 "이 상품을 본 고객이 함께 구매한 상품" |

#### 아이템 기반 협업 필터링 과정

1. **상품-유저 평점 매트릭스** 만들기 (유저 기반 매트릭스를 뒤집은 것 = 전치)
2. **상품 간 코사인 유사도 계산** (모든 상품 쌍의 유사도를 구함)
3. **특정 상품과 가장 유사한 상품 Top N** = "이 상품을 보신 분은 이런 상품도 좋아합니다"

> 아이템 기반이 실무에서 더 많이 쓰이는 이유: 유저의 취향은 시간에 따라 변하지만,
> 상품 간의 관계(예: 노트북-마우스)는 잘 변하지 않아서 추천이 더 안정적입니다.
