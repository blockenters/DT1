### 개념 설명: 클러스터링이란?

**클러스터링 = 비슷한 것끼리 자동으로 묶기**

클러스터링은 **비지도학습(Unsupervised Learning)**의 대표적인 기법입니다.
비지도학습이란? **정답(라벨) 없이** 데이터 속에서 스스로 패턴을 발견하는 것입니다.

#### 지도학습(분류) vs 비지도학습(클러스터링) 차이

| 구분 | 지도학습 (분류) | 비지도학습 (클러스터링) |
|------|----------------|----------------------|
| 정답 유무 | 있음 (라벨) | 없음 |
| 예시 | "이 고객은 이탈이야" | "비슷한 고객끼리 묶어줘" |
| 목적 | 정답을 예측 | 스스로 그룹을 발견 |

#### K-Means 알고리즘 동작 원리 (쉽게 이해하기)

K-Means는 가장 많이 쓰이는 클러스터링 알고리즘입니다. 동작 방식은 의외로 단순합니다:

1. **K개의 중심점을 랜덤으로 놓음** (K=3이면 3개의 점을 아무데나 찍음)
2. **각 데이터를 가장 가까운 중심점에 배정** (가까운 중심점의 그룹에 소속)
3. **각 그룹의 평균 위치로 중심점을 이동** (그룹원들의 한가운데로)
4. **2~3을 반복** → 더 이상 중심점이 변하지 않으면 완료!

> **주의: 스케일링 필수!**
>
> 클러스터링은 데이터 간 "거리"를 기준으로 그룹을 나눕니다.
> 예를 들어 키 170cm vs 몸무게 70kg → 단위가 다르면 거리 계산이 왜곡됩니다.
> 키의 차이 10cm와 몸무게의 차이 10kg은 전혀 다른 의미인데, 숫자만 보면 같아 보이죠.
> 그래서 **모든 피처를 같은 스케일로 맞추는 스케일링이 필수**입니다!

### 개념 설명: 최적의 K는 어떻게 찾을까?

K-Means에서는 **K(클러스터 수)를 우리가 미리 정해야** 합니다.
그런데 고객을 3그룹으로 나눌지, 5그룹으로 나눌지 어떻게 알 수 있을까요?

#### 1. 엘보우 기법 (Elbow Method)
- K를 2, 3, 4, 5... 늘려가면서 **"그룹 내 거리 합(inertia)"**을 계산합니다
- 이 값을 그래프로 그리면 → **팔꿈치처럼 꺾이는 지점**이 최적의 K입니다
- 왜? K를 늘리면 거리 합이 줄어들지만, 어느 순간부터 줄어드는 폭이 확 줄어듭니다

#### 2. 실루엣 점수 (Silhouette Score)
- **-1 ~ 1 사이의 점수**로, 클러스터가 얼마나 잘 분리되었는지를 나타냅니다
- **1에 가까울수록** → 클러스터가 잘 나뉨 (좋음!)
- **0에 가까울수록** → 클러스터 경계가 모호
- **가장 높은 실루엣 점수를 가진 K를 선택**합니다

#### 현업에서는?
- 두 방법을 **모두 확인**합니다 (결과가 다를 수 있음)
- 추가로 **비즈니스 관점에서 해석 가능한 K**인지도 고려합니다
- 예: 통계적으로 K=7이 최적이어도, 마케팅팀이 7개 등급을 관리하기 어려우면 K=4로 줄이기도 합니다

### 개념 설명: RFM 분석이란?

**RFM = 고객 가치를 평가하는 현업 표준 프레임워크**

고객을 분석할 때, 가장 널리 쓰이는 3가지 지표의 앞글자를 따서 RFM이라고 합니다:

| 지표 | 뜻 | 설명 | 좋은 방향 |
|------|-----|------|----------|
| **R** (Recency) | 최근성 | 마지막 구매가 얼마나 최근인가? | 작을수록 좋음 (최근에 샀다) |
| **F** (Frequency) | 빈도 | 얼마나 자주 구매하는가? | 클수록 좋음 (자주 산다) |
| **M** (Monetary) | 금액 | 얼마나 많이 썼는가? | 클수록 좋음 (많이 산다) |

#### 왜 RFM을 쓰는가?
모든 고객에게 **똑같은 마케팅**을 하면 → **비효율적**입니다!
RFM으로 고객을 그룹화하면 → **각 그룹에 맞는 맞춤형 전략**을 세울 수 있습니다.

#### RFM 기반 고객 세그먼트 예시

| 세그먼트 | R (최근성) | F (빈도) | M (금액) | 마케팅 전략 |
|---------|-----------|---------|---------|-----------|
| **VIP** | 낮음 (최근 구매) | 높음 (자주) | 높음 (많이) | 감사 혜택, VIP 전용 이벤트 |
| **이탈위험** | 높음 (오래 전) | 낮음 (드물게) | 낮음 (조금) | 복귀 쿠폰, 긴급 할인 |
| **신규** | 낮음 (최근 가입) | 낮음 (아직 소수) | 낮음 (아직 소수) | 온보딩 프로모션, 첫 구매 혜택 |

이제 실제 데이터에 K-Means를 적용해서 RFM 기반 고객 세분화를 해봅시다!
