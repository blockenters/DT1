## 2. 로지스틱 회귀 (Logistic Regression)

이름은 "회귀"지만 **분류 모델**입니다!

**핵심 아이디어:** 데이터를 보고 **확률(0~1)**을 계산해서, 0.5 이상이면 이탈(1), 미만이면 유지(0)로 분류합니다.

예: `이탈 확률 = 0.82` → 이탈로 예측!

---
## 분류와 로지스틱 회귀를 배우기 전에

### 분류(Classification)란?

**분류 = 카테고리(범주)를 예측하는 것**

| 예측 대상 | 결과 예시 | 유형 |
|----------|----------|------|
| 고객 이탈 여부 | 이탈 / 유지 | 분류 (카테고리) |
| 메일 종류 | 스팸 / 정상 | 분류 (카테고리) |
| 거래 유형 | 사기 / 정상 | 분류 (카테고리) |
| 아파트 가격 | 3억 5천만원 | 회귀 (숫자) ← 지난 시간에 배움! |

### 로지스틱 회귀란?

**이름은 "회귀"지만 실제로는 분류 모델입니다!**

로지스틱 회귀는 **0과 1 사이의 확률**로 예측합니다.

```
모델 → "이 고객의 이탈 확률은 0.82 (82%)" → 이탈로 예측!
모델 → "이 고객의 이탈 확률은 0.15 (15%)" → 유지로 예측!
```

### 임계값(Threshold)이란?

확률을 카테고리로 바꾸는 **기준점**입니다. 기본값은 **0.5**입니다.

| 이탈 확률 | 임계값 0.5 기준 | 예측 결과 |
|----------|---------------|----------|
| 0.82 | 0.82 > 0.5 | **이탈 (1)** |
| 0.45 | 0.45 ≤ 0.5 | **유지 (0)** |
| 0.51 | 0.51 > 0.5 | **이탈 (1)** |

### 로지스틱 회귀의 장단점

| 장점 | 단점 |
|------|------|
| **확률**을 알 수 있음 | 복잡한 패턴은 못 잡음 |
| 해석이 쉬움 | 비선형 관계에 약함 |
| 빠르고 안정적 | 피처 엔지니어링 필요 |

### 현업 활용 예시

> "이 고객은 **73% 확률로 이탈**할 것으로 예측됩니다"
> → 확률이 높은 고객부터 **우선순위**를 정해서 관리!
> → 상위 10% 고객에게 쿠폰 발송, 상담 전화 등 **선제 대응**

## 3. 결정트리 (Decision Tree)

**if-else 규칙**으로 데이터를 나누어 분류합니다.

- 사람이 이해하기 쉬운 모델
- 예: "최근접속일수 > 10이고 불만접수 > 2이면 → 이탈"
- 단점: 과적합(overfitting)되기 쉬움

---
## 결정트리를 배우기 전에

### 결정트리(Decision Tree)란?

**결정트리 = 스무고개 게임과 같습니다!**

질문을 계속해서 데이터를 분류합니다:

```
Q1. 최근 접속이 30일 이상 지났나?
├── Yes → Q2. 불만 접수가 2회 이상인가?
│         ├── Yes → 이탈! (확률 높음)
│         └── No  → 유지 (접속은 안 했지만 불만은 없음)
└── No  → Q3. 월 이용금액이 5만원 이상인가?
          ├── Yes → 유지 (활발하게 사용 중)
          └── No  → 이탈 위험 (접속은 하지만 사용량 적음)
```

### 결정트리의 장단점

| 장점 | 단점 |
|------|------|
| **사람이 이해하기 가장 쉬움** | **과적합**되기 쉬움 |
| 시각화가 가능 (트리 그림) | 데이터가 조금만 바뀌어도 트리가 크게 변함 |
| 스케일링 불필요 | 단일 트리는 성능이 낮을 수 있음 |
| 피처 중요도를 알 수 있음 | - |

### 과적합을 방지하려면?

`max_depth` 파라미터로 트리의 깊이를 제한합니다.
- `max_depth=3` → 질문을 최대 3번만 함 (간단한 트리)
- `max_depth=10` → 질문을 10번까지 함 (복잡한 트리 → 과적합 위험)
- `max_depth=None` → 제한 없음 (과적합 위험 매우 높음)

> **팁:** 결정트리 단독으로 쓰기보다, 이것을 여러 개 합친 **랜덤포레스트**나 **XGBoost**를 더 많이 사용합니다!

## 6. 분류 평가지표

분류 모델의 성능을 측정하는 5가지 지표:

| 지표 | 의미 | 언제 쓰나? |
|------|------|------------|
| **정확도 (Accuracy)** | 전체 중 맞은 비율 | 데이터가 균형일 때 |
| **정밀도 (Precision)** | 이탈이라 예측한 것 중 실제 이탈 비율 | 잘못된 알림을 줄이고 싶을 때 |
| **재현율 (Recall)** | 실제 이탈 중 잡아낸 비율 | 이탈을 놓치면 안 될 때 |
| **F1 Score** | 정밀도와 재현율의 균형 | 둘 다 중요할 때 |
| **AUC** | 모델의 전반적 분류 능력 (0.5~1.0) | 모델 전체 성능 비교 |

---
## 분류 평가지표를 배우기 전에

### 왜 정확도(Accuracy)만 보면 안 될까요?

**충격적인 예시:**

고객 1,000명 중 이탈 고객이 50명(5%)인 데이터가 있다고 합시다.
모델이 **"모든 고객은 유지입니다"**라고 예측하면?

→ 정확도 = 950/1000 = **95%!!**

정확도는 95%로 높지만, **이탈 고객을 한 명도 못 잡았습니다!**
이런 모델은 쓸모가 없겠죠?

### 혼동행렬 (Confusion Matrix)

모든 분류 평가지표의 기본이 되는 표입니다:

```
                  예측: 유지(0)    예측: 이탈(1)
실제: 유지(0)      TN (정확!)       FP (오탐 - 멀쩡한데 이탈이라 함)
실제: 이탈(1)      FN (놓침!!)      TP (정확!)
```

| 약어 | 의미 | 설명 |
|------|------|------|
| **TP** (True Positive) | 이탈을 이탈로 맞춤 | 잘 잡음! |
| **TN** (True Negative) | 유지를 유지로 맞춤 | 잘 맞춤! |
| **FP** (False Positive) | 유지인데 이탈이라 함 | 오탐 (괜히 알림) |
| **FN** (False Negative) | 이탈인데 유지라 함 | **놓침! (가장 위험)** |

### 핵심 평가지표 4가지

#### 1. 정밀도 (Precision)
- **"이탈이라 예측한 것 중 실제 이탈 비율"**
- = TP / (TP + FP)
- 비유: "이탈 알림의 정확도" → 알림을 보냈는데 실제로 이탈한 비율
- 정밀도가 낮으면? → 멀쩡한 고객에게 불필요한 쿠폰/전화를 보냄

#### 2. 재현율 (Recall)
- **"실제 이탈 중 잡아낸 비율"**
- = TP / (TP + FN)
- 비유: "이탈 고객을 얼마나 놓치지 않는가" → 그물의 촘촘함
- 재현율이 낮으면? → 이탈할 고객을 놓쳐서 고객을 잃음

#### 3. F1 Score
- **정밀도와 재현율의 균형** (둘의 조화평균)
- 정밀도와 재현율 둘 다 중요할 때 사용
- 0~1 사이, 1에 가까울수록 좋음

#### 4. AUC (Area Under the Curve)
- **모델의 전반적 분류 능력**
- 0.5 = 동전 던지기(찍기) ~ 1.0 = 완벽한 분류
- 임계값에 관계없이 모델 자체의 성능을 평가

### 현업에서의 선택 기준

| 상황 | 중요한 지표 | 이유 |
|------|-----------|------|
| **고객 이탈 예측** | 재현율 | 놓치면 고객을 잃음 (FN이 치명적) |
| **사기 탐지** | 정밀도 + 재현율 둘 다 | 놓쳐도 안 되고, 오탐해도 고객 불편 |
| **스팸 필터** | 정밀도 | 정상 메일을 스팸으로 분류하면 중요한 메일을 못 봄 |
| **질병 진단** | 재현율 | 환자를 정상이라고 하면 치료 기회를 놓침 |
