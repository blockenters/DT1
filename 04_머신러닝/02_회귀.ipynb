## 2. 선형회귀 (Linear Regression)

가장 기본적인 회귀 모델입니다.

**핵심 아이디어:** 데이터를 가장 잘 설명하는 **직선(또는 평면)** 을 찾아서 예측합니다.

예: `매매가 = a × 면적 + b × 층 + c × ... + 절편`

---
## 선형회귀를 배우기 전에

### 회귀(Regression)란?

**회귀 = 연속적인 숫자를 예측하는 것**

| 예측 대상 | 결과 예시 | 유형 |
|----------|----------|------|
| 아파트 가격 | 3억 5천만원 | 회귀 (숫자) |
| 다음 달 매출 | 1,200만원 | 회귀 (숫자) |
| 내일 기온 | 23.5도 | 회귀 (숫자) |
| 스팸 여부 | 스팸/정상 | 분류 (카테고리) ← 이건 다음 시간! |

### 선형회귀의 핵심 아이디어

**데이터를 가장 잘 설명하는 직선(또는 평면)을 찾는 것!**

중학교 수학에서 배운 `y = ax + b`를 기억하시나요?
- **a (기울기)** = x가 1 증가하면 y가 얼마나 변하는지
- **b (절편)** = x가 0일 때 y의 값

머신러닝에서는 여러 변수를 사용합니다:
```
매매가 = a1 × 면적 + a2 × 층수 + a3 × 건물나이 + ... + b(절편)
```

모델이 하는 일 = 데이터를 보고 **a1, a2, a3, ..., b 값을 찾는 것!**

### 선형회귀의 장단점

| 장점 | 단점 |
|------|------|
| 해석이 쉬움 (a가 크면 영향이 큼) | 복잡한 관계는 못 잡음 (직선으로만 예측) |
| 학습 속도가 매우 빠름 | 이상치(outlier)에 민감함 |
| 결과를 설명하기 좋음 | 변수 간 비선형 관계를 놓침 |

> **비유:** 선형회귀는 자(ruler)로 직선을 긋는 것. 데이터가 직선에 가까우면 잘 맞지만, 곡선이면 한계가 있습니다.

## 3. 랜덤포레스트 회귀 (Random Forest)

여러 개의 결정트리(Decision Tree)를 만들고, 각 트리의 예측값을 **평균**내서 최종 예측합니다.

- 성능이 좋고 안정적
- 과적합(overfitting)에 강함
- 스케일링 없이도 잘 동작

---
## 랜덤포레스트를 배우기 전에

### 결정트리(Decision Tree)란?

**결정트리 = if-else 규칙의 나무 구조**

컴퓨터가 스무고개처럼 질문을 만들어서 예측합니다:

```
Q1. 면적이 50㎡ 보다 큰가?
├── Yes → Q2. 층수가 10층 이상인가?
│         ├── Yes → 가격: 6억
│         └── No  → 가격: 4억
└── No  → Q3. 역세권인가?
          ├── Yes → 가격: 3억
          └── No  → 가격: 2억
```

### 랜덤포레스트(Random Forest)란?

**랜덤포레스트 = 여러 개의 결정트리를 만들어서 평균내기 (앙상블)**

| 비유 | 설명 |
|------|------|
| 한 명에게 물어보기 | 결정트리 1개 → 그 사람이 틀리면 끝 |
| 100명에게 물어서 평균내기 | 랜덤포레스트 → 여러 의견의 평균이라 안정적 |

### 왜 좋은가요?

단일 결정트리는 **과적합(overfitting)**되기 쉽습니다.
- 과적합 = 학습 데이터는 완벽히 맞추지만, 새로운 데이터에서는 틀리는 것
- 비유: 교과서 내용은 달달 외웠는데, 응용문제는 못 푸는 것

랜덤포레스트는 여러 트리의 평균을 내기 때문에 **과적합에 강합니다!**

| 장점 | 단점 |
|------|------|
| 성능이 좋음 | 해석이 어려움 (트리가 100개!) |
| 과적합에 강함 | 선형회귀보다 느림 |
| 피처 중요도를 알 수 있음 | 메모리를 많이 사용 |
| 스케일링 불필요 | - |

## 4. XGBoost 회귀

현업(캐글, 기업)에서 **가장 많이 쓰는 모델**입니다.

- 트리를 순차적으로 만들면서 이전 트리의 **오차를 보완**
- 속도가 빠르고 성능이 뛰어남
- 대회 우승 모델의 단골

---
## XGBoost를 배우기 전에

### XGBoost란?

**XGBoost = 현업에서 가장 많이 쓰는 모델!**

캐글(Kaggle) 대회에서 우승한 모델의 상당수가 XGBoost를 사용했습니다.

### Gradient Boosting의 핵심 아이디어

**"이전 트리의 실수를 다음 트리가 보완하는 방식"**

| 단계 | 설명 | 비유 |
|------|------|------|
| 1번째 트리 | 전체 데이터를 대략 예측 | 시험 1회차: 60점 |
| 2번째 트리 | 1번째 트리가 틀린 부분을 집중 학습 | 틀린 문제 오답노트 → 75점 |
| 3번째 트리 | 2번째까지도 틀린 부분을 집중 학습 | 또 오답노트 → 85점 |
| ... | 계속 반복 | 점점 점수가 올라감! |

### 랜덤포레스트 vs XGBoost 비교

| 구분 | 랜덤포레스트 | XGBoost |
|------|-------------|---------|
| **방식** | 여러 트리를 **동시에(병렬)** 만들고 평균 | 트리를 **순서대로(순차)** 만들며 개선 |
| **비유** | 100명이 동시에 시험 → 평균 | 1명이 시험 → 오답노트 → 재시험 → 반복 |
| **성능** | 좋음 | 대부분 더 좋음 |
| **속도** | 빠름 (병렬 처리) | 상대적으로 느림 (순차 처리) |
| **튜닝** | 쉬움 (파라미터 적음) | 어려움 (파라미터 많음) |

### XGBoost의 장단점

| 장점 | 단점 |
|------|------|
| 대부분의 경우 최고 성능 | 하이퍼파라미터가 많음 |
| 결측값 자동 처리 | 학습이 상대적으로 느림 |
| 과적합 방지 기능 내장 | 해석이 어려움 |
| 캐글 대회 우승 모델 다수 | 데이터가 적으면 과적합 위험 |

## 5. 모델 평가

회귀 모델의 성능을 측정하는 3가지 지표:

| 지표 | 의미 | 언제 쓰나? |
|------|------|------------|
| **MAE** (평균절대오차) | 예측이 평균적으로 얼마나 틀리는지 | 직관적으로 오차 크기 파악 |
| **RMSE** (평균제곱근오차) | 큰 오차에 더 큰 벌점 | 큰 오차가 치명적일 때 |
| **R²** (결정계수) | 모델이 데이터를 얼마나 잘 설명하는지 (0~1) | 모델 전체 성능 비교 |

---
## 모델 평가를 배우기 전에

### 모델을 만들었는데, 얼마나 좋은지 어떻게 알까요?

회귀 모델은 **숫자를 예측**하므로, "예측값이 실제값과 얼마나 차이나는가"를 측정합니다.

### 3가지 핵심 평가지표

#### 1. MAE (Mean Absolute Error, 평균 절대 오차)
- **예측값과 실제값 차이의 평균**
- 가장 직관적! → "이 모델은 평균적으로 ±500만원 틀립니다"
- 예: MAE = 500이면, 평균적으로 500만원 정도 오차

#### 2. RMSE (Root Mean Squared Error, 평균 제곱근 오차)
- MAE와 비슷하지만, **큰 오차에 더 민감**
- 예: 500만원 오차 10개보다, 5000만원 오차 1개가 RMSE를 더 크게 만듦
- 이상치(outlier)가 중요한 문제에서 사용

#### 3. R² (R-Squared, 결정계수)
- **0부터 1 사이의 값** (1에 가까울수록 좋음)
- "모델이 데이터 변동의 몇 %를 설명하는가?"
- 예: R² = 0.85 → "이 모델은 가격 변동의 85%를 설명합니다"

### 현업에서의 기준 (참고)

| R² 값 | 해석 |
|--------|------|
| 0.9 이상 | 매우 좋은 모델 |
| 0.7 ~ 0.9 | 괜찮은 모델 |
| 0.5 ~ 0.7 | 보통 (개선 필요) |
| 0.5 미만 | 모델을 다시 만들어야 함 |

> **팁:** MAE/RMSE는 **낮을수록** 좋고, R²는 **높을수록** 좋습니다!
