## 학습목표
- Pipeline으로 전처리~모델링을 하나로 묶는다
- ColumnTransformer로 숫자/범주형 피처를 한번에 처리한다
- 모델을 저장하고 로드하여 재사용한다
- 실무에서 주의할 점을 이해한다

## 왜 배워야 하나요?

실무에서는 전처리, 학습, 예측을 **따로따로** 하면 실수가 생깁니다.

| 문제 | Pipeline 해결 |
|------|---------------|
| 전처리 코드를 다시 쓰기 귀찮음 | 한번 만들면 재사용 |
| 학습/테스트에 다른 전처리 적용 | 자동으로 일관성 유지 |
| 모델 배포할 때 전처리 빠뜨림 | 전처리+모델을 한 파일로 저장 |

---
### 개념 설명: 왜 Pipeline이 필요한가?

#### 문제: 전처리와 모델을 따로따로 하면 생기는 문제

머신러닝에서는 보통 이런 단계를 거칩니다:
1. 데이터 스케일링 (전처리)
2. 모델 학습
3. 새 데이터에 스케일링 적용 (전처리)
4. 예측

이걸 **따로따로** 코드를 쓰면:
- 코드가 길어지고 복잡해짐
- 실수로 테스트 데이터에 스케일링을 빠뜨릴 수 있음
- 새로운 데이터에 적용할 때 전처리 코드를 다시 써야 함

#### Pipeline = 전처리 + 모델을 하나로 묶는 것

```python
# Pipeline 없이 (따로따로)           # Pipeline으로 (한번에)
scaler.fit(X_train)                  pipe.fit(X_train, y_train)
X_train_scaled = scaler.transform()  pipe.predict(X_test)
model.fit(X_train_scaled)            # 끝! 스케일링+예측 자동!
X_test_scaled = scaler.transform()
model.predict(X_test_scaled)
```

#### Pipeline의 3가지 장점
1. **코드 깔끔** - 전처리+학습+예측이 각각 한 줄
2. **실수 방지** - 스케일링을 빠뜨리거나 잘못 적용할 일이 없음
3. **모델 저장할 때 전처리도 같이 저장됨** - 배포할 때 매우 편리!

### 개념 설명: ColumnTransformer란?

#### 문제: 컬럼마다 다른 전처리가 필요하다

실제 데이터에는 여러 종류의 컬럼이 섞여 있습니다:
- **숫자 컬럼** (면적, 가격, 나이 등) → **스케일링(StandardScaler)** 필요
- **범주형 컬럼** (지역, 성별, 등급 등) → **인코딩(OneHotEncoder)** 필요

이걸 각각 따로 처리하면 코드가 복잡해지고 실수하기 쉽습니다.

#### ColumnTransformer = 컬럼별로 다른 전처리를 한번에 지정

```python
ColumnTransformer([
    ('숫자처리', StandardScaler(), ['면적', '층', '나이']),     # 숫자 → 스케일링
    ('범주처리', OneHotEncoder(), ['지역', '건물유형']),         # 범주 → 인코딩
])
```

이렇게 만든 ColumnTransformer를 **Pipeline에 넣으면**:
- 원본 데이터(숫자+범주 섞인 상태)를 바로 넣어도
- 자동으로 숫자는 스케일링, 범주는 인코딩 처리 후 모델 학습!

> 실무에서는 거의 항상 ColumnTransformer + Pipeline 조합을 사용합니다.

### 개념 설명: 왜 모델을 저장해야 하는가?

#### 문제: 학습에 시간이 오래 걸리는 모델을 매번 다시 학습하면 비효율적

대용량 데이터로 모델을 학습하면 몇 시간~며칠이 걸릴 수 있습니다.
매번 다시 학습하는 것은 시간과 비용 낭비입니다.

#### joblib: Python 객체를 파일로 저장/로드

- `joblib.dump(모델, "파일명.pkl")` → 모델을 파일로 저장
- `joblib.load("파일명.pkl")` → 저장된 모델을 불러오기
- pickle이라는 기본 도구보다 **대용량 데이터에 더 효율적**입니다

#### 현업에서의 모델 배포 흐름

```
1. 데이터 수집 & 전처리 & 모델 학습  (데이터 사이언티스트)
2. 학습된 모델을 .pkl 파일로 저장     (joblib.dump)
3. 서버에 .pkl 파일 배포              (엔지니어)
4. API로 실시간 예측 서비스 제공       (joblib.load → predict)
```

> **중요:** 모델을 저장할 때 **전처리 파이프라인도 같이 저장**해야 합니다!
> 모델만 저장하면, 새 데이터에 전처리를 적용하는 코드를 또 따로 관리해야 합니다.
> **Pipeline으로 묶어서 저장하면** 전처리+모델이 한 파일에 들어가므로 관리가 훨씬 쉽습니다.
> 이것이 바로 Pipeline을 사용하는 가장 큰 이유 중 하나입니다!

## 5. 실무 적용 팁

### 데이터 누수 (Data Leakage) 주의!
- 테스트 데이터의 정보가 학습에 포함되면 안 됩니다
- 스케일링할 때 **학습 데이터로만 fit** 해야 합니다
- Pipeline을 쓰면 자동으로 방지됩니다!

### 검증 전략
- 시계열 데이터는 시간순으로 분리 (미래 데이터로 학습하면 안 됨)
- 클래스 불균형이 심하면 Stratified K-Fold 사용

### 모델 모니터링
- 배포 후에도 성능이 유지되는지 주기적으로 확인
- 데이터 분포가 바뀌면 모델을 재학습해야 합니다

### 정리

| 단계 | 도구 |
|------|------|
| 전처리 자동화 | ColumnTransformer |
| 전처리+모델 묶기 | Pipeline |
| 하이퍼파라미터 탐색 | GridSearchCV |
| 안정적 평가 | cross_val_score |
| 모델 저장/로드 | joblib |
