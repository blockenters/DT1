### 개념 설명: 왜 교차검증이 필요한가?

#### 문제: 한 번만 나누면 "운"에 따라 결과가 달라짐

데이터를 train/test로 **한 번만** 나누면, 어떤 데이터가 테스트에 들어갔느냐에 따라 결과가 크게 달라질 수 있습니다.
운 좋게 쉬운 데이터가 테스트에 들어가면 점수가 높고, 어려운 데이터가 들어가면 낮아지죠.

#### 교차검증(Cross Validation) = K번 나눠서 평균내기

데이터를 **K등분**해서, 각각 한 번씩 테스트 세트로 사용합니다. K번 평가한 후 **평균**을 냅니다.

**5-Fold 교차검증 예시:**
```
1회차: [테스트] [학습] [학습] [학습] [학습]  → 점수1
2회차: [학습] [테스트] [학습] [학습] [학습]  → 점수2
3회차: [학습] [학습] [테스트] [학습] [학습]  → 점수3
4회차: [학습] [학습] [학습] [테스트] [학습]  → 점수4
5회차: [학습] [학습] [학습] [학습] [테스트]  → 점수5

최종 점수 = (점수1 + 점수2 + 점수3 + 점수4 + 점수5) / 5
```

#### 장점
- **모든 데이터가 한 번씩 테스트에 사용됨** → 운의 영향 최소화
- 결과의 **신뢰도가 훨씬 높음**
- 표준편차를 보면 **모델이 얼마나 안정적인지**도 알 수 있음

> **현업에서:** 모델 성능을 보고할 때 반드시 교차검증 결과를 사용합니다.
> "이 모델의 정확도는 85%입니다"가 아니라 "5-Fold CV 평균 정확도 85% (+-2%)"처럼 보고합니다.

### 개념 설명: 하이퍼파라미터와 GridSearchCV

#### 하이퍼파라미터란?
모델이 **스스로 학습하는 값이 아니라**, 우리가 **직접 설정해주는 값**입니다.

| 하이퍼파라미터 | 의미 | 예시 |
|--------------|------|------|
| n_estimators | 트리를 몇 개 만들 것인가 | 50, 100, 200 |
| max_depth | 트리의 최대 깊이 | 3, 5, 10, 무제한 |
| learning_rate | 학습 속도 (한 번에 얼마나 배울지) | 0.01, 0.1, 0.3 |

이 값들을 어떻게 설정하느냐에 따라 모델 성능이 크게 달라집니다!

#### GridSearchCV = 최적의 하이퍼파라미터를 자동으로 찾아줌

일일이 바꿔가며 테스트하는 건 너무 번거롭죠. GridSearchCV는 **모든 조합을 자동으로 시도**합니다.

예를 들어:
- n_estimators = [50, 100, 200] (3가지)
- max_depth = [3, 5, 10] (3가지)
- → **3 x 3 = 9가지 조합**을 자동으로 전부 시도!

게다가 내부적으로 **교차검증까지** 해주기 때문에, 가장 **신뢰할 수 있는 최적 파라미터**를 찾아줍니다.

> Grid(격자)라는 이름은 파라미터 조합을 격자처럼 모두 탐색한다는 의미입니다.

### 개념 설명: 과적합과 과소적합 이해하기

#### 과적합 (Overfitting) - "시험 범위만 달달 외움"
- 모델이 학습 데이터를 **"외워버린"** 상태입니다
- 학습 데이터에서는 완벽하지만, **새로운 데이터에서 성능이 급락**합니다
- **증상:** 학습 점수 99%, 테스트 점수 60% → 차이가 크면 과적합!
- **비유:** 수학 시험에서 기출문제만 달달 외움 → 새로운 유형의 문제가 나오면 못 풀음
- **해결법:**
  - max_depth 줄이기 (모델을 단순하게)
  - 데이터 늘리기 (더 많은 사례를 학습)
  - 규제(Regularization) 적용 (모델이 너무 복잡해지지 않게 제한)

#### 과소적합 (Underfitting) - "아예 공부를 안 함"
- 모델이 학습 자체를 **제대로 못한** 상태입니다
- 학습 데이터에서도, 테스트 데이터에서도 **성능이 낮습니다**
- **증상:** 학습 점수 50%, 테스트 점수 45% → 둘 다 낮음
- **비유:** 아예 공부를 안 하고 시험을 봄 → 모든 문제를 못 풀음
- **해결법:**
  - 더 복잡한 모델 사용 (선형회귀 → 랜덤포레스트)
  - 피처(변수) 추가
  - max_depth 늘리기 (모델이 더 깊이 학습하게)

#### 핵심: 학습 점수와 테스트 점수의 차이를 보면 알 수 있다!

| 상황 | 학습 점수 | 테스트 점수 | 진단 |
|------|----------|-----------|------|
| 학습 높고, 테스트 낮음 | 99% | 60% | 과적합 |
| 둘 다 낮음 | 50% | 45% | 과소적합 |
| 둘 다 높고 비슷함 | 90% | 87% | 적절한 학습 (Good!) |

### 개념 설명: 피처 중요도란?

**피처 중요도(Feature Importance) = "어떤 변수가 예측에 가장 영향을 많이 미치는가?"**

예를 들어, 집값을 예측하는 모델에서:
- "면적"이 가장 중요한 변수일 수도 있고
- "역세권(분)"이 가장 중요한 변수일 수도 있습니다
- 피처 중요도를 보면 이걸 **숫자로 정확히** 알 수 있습니다!

#### 왜 피처 중요도가 중요한가?

1. **불필요한 피처 제거** → 모델이 단순해지고, 학습 속도가 빨라집니다
2. **비즈니스 인사이트 발견** → 예: "고객 이탈에 가장 큰 영향을 주는 요인 = 최근접속일수"
   - 이 정보를 바탕으로 "오래 접속하지 않은 고객에게 알림 보내기" 같은 대응 전략을 세울 수 있습니다
3. **모델 신뢰성 확인** → 상식적으로 중요하지 않은 변수가 1위라면 모델에 문제가 있을 수 있습니다

> **트리 기반 모델(랜덤포레스트, XGBoost 등)은 학습이 끝나면 자동으로 피처 중요도를 제공합니다.**
> `.feature_importances_` 속성으로 바로 확인할 수 있어서 매우 편리합니다!
