## 1. 머신러닝이란?

**사람이 규칙을 하나하나 정하는 대신, 데이터를 주면 컴퓨터가 스스로 패턴을 찾아내는 것**

### 머신러닝의 종류

| 종류 | 설명 | 예시 | 실무 활용 |
|------|------|------|----------|
| **지도학습 - 회귀** | 숫자를 예측 | 집값, 매출, 온도 | 부동산 가격 예측, 재고 수요 예측 |
| **지도학습 - 분류** | 카테고리를 예측 | 스팸/정상, 합격/불합격 | 고객 이탈 예측, 불량품 검출 |
| **비지도학습 - 클러스터링** | 비슷한 것끼리 묶기 | 고객 그룹 나누기 | 고객 세분화, 이상 탐지 |
| **추천 시스템** | 좋아할 것 추천 | 넷플릭스, 쿠팡 | 상품 추천, 콘텐츠 추천 |

### ML 워크플로우

```
데이터 수집 → 전처리(정리) → 모델 학습 → 예측 → 평가
```

1. **데이터 수집**: DB, CSV, API 등에서 데이터를 모은다
2. **전처리**: 결측값 처리, 인코딩, 스케일링 등 데이터를 깨끗하게 정리
3. **모델 학습**: 알고리즘에 데이터를 넣어서 패턴을 학습시킴
4. **예측**: 새로운 데이터에 대해 결과를 예측
5. **평가**: 예측이 얼마나 정확한지 확인

## 머신러닝을 배우기 전에 꼭 알아야 할 것들

### 머신러닝이란 무엇인가요?

**머신러닝(Machine Learning)**은 한마디로 **데이터에서 패턴을 찾아 예측하는 기술**입니다.

우리가 수학 문제를 많이 풀면 새로운 문제도 풀 수 있게 되듯이,
컴퓨터도 데이터를 많이 보면 새로운 데이터에 대해 예측할 수 있게 됩니다.

### 전통 프로그래밍 vs 머신러닝

| 구분 | 전통 프로그래밍 | 머신러닝 |
|------|----------------|----------|
| **방식** | 사람이 규칙을 직접 작성 | 데이터에서 규칙을 자동으로 학습 |
| **예시** | "가격이 3만원 이상이면 비싸다" | 데이터를 보고 "비싸다"의 기준을 스스로 찾음 |
| **장점** | 규칙이 명확할 때 정확 | 규칙이 복잡하거나 모를 때 유용 |
| **단점** | 규칙이 많아지면 관리 불가 | 데이터가 충분해야 함 |

**쉬운 비유:**
- 전통 프로그래밍 = 요리 레시피를 보고 그대로 따라 만드는 것
- 머신러닝 = 맛있는 음식을 많이 먹어보고, 어떤 재료 조합이 맛있는지 스스로 깨닫는 것

### 머신러닝의 종류

| 종류 | 한줄 설명 | 쉬운 비유 | 실생활 예시 |
|------|----------|----------|------------|
| **지도학습 - 회귀** | 숫자를 예측 | 수학 시험 점수 예측 | 아파트 가격 예측, 매출 예측 |
| **지도학습 - 분류** | 카테고리를 예측 | O/X 퀴즈 맞추기 | 스팸 메일 분류, 고객 이탈 예측 |
| **비지도학습** | 비슷한 것끼리 묶기 | 비슷한 친구끼리 모이는 것 | 고객 그룹 나누기, 이상 거래 탐지 |
| **추천 시스템** | 좋아할 것을 추천 | 친구가 "너 이거 좋아할 걸?" 하는 것 | 넷플릭스 영화 추천, 유튜브 알고리즘 |

## 3. 데이터 분리 (Train/Test Split)

### 왜 데이터를 나누나요?

시험에 비유하면:
- **학습 데이터(Train)** = 교과서, 문제집 → 공부할 때 사용
- **테스트 데이터(Test)** = 실제 시험 → 실력을 확인할 때 사용

시험 문제를 미리 보고 공부하면 100점 맞겠지만, **진짜 실력은 아닙니다.**  
마찬가지로, 학습에 사용한 데이터로 평가하면 모델이 얼마나 잘하는지 알 수 없습니다.

### 실무에서는?
- 보통 **80% 학습, 20% 테스트**로 나눔
- `random_state`를 고정하면 매번 같은 결과를 얻을 수 있음 (재현성)

---
## 데이터를 왜 나눠야 할까요?

### 시험 비유로 이해하기

학교 시험을 생각해보세요:
- **시험 문제를 미리 보고 공부하면?** → 100점 맞겠지만, 진짜 실력은 아닙니다
- **새로운 문제로 시험을 봐야** → 진짜 실력을 알 수 있습니다

머신러닝도 똑같습니다!

| 데이터 종류 | 비유 | 비율 | 용도 |
|------------|------|------|------|
| **학습 데이터 (Train)** | 교과서, 문제집 | 80% | 모델이 패턴을 학습하는 데 사용 |
| **테스트 데이터 (Test)** | 실제 시험 | 20% | 모델의 진짜 실력을 평가하는 데 사용 |

### random_state란?

`random_state=42`를 설정하면 **매번 같은 방식으로 데이터를 나눕니다.**

- 비유: "같은 시험 문제로 시험 보기" → 결과를 비교할 수 있음
- 숫자 42에 특별한 의미는 없습니다 (관례적으로 42를 많이 씀)
- 이것을 **재현성(Reproducibility)**이라고 합니다 → 누가 실행해도 같은 결과!

## 4. 피처 엔지니어링

### 피처(Feature)란?

모델에게 주는 **정보(입력값)**를 피처라고 합니다.  
좋은 피처를 만들면 모델 성능이 크게 올라갑니다.

예를 들어:
- 건축년도 자체보다 **건물 나이**가 더 유용한 정보
- 면적과 가격을 합쳐서 **면적당 가격**을 만들면 비교가 쉬움

### 실무에서는?
- 모델 성능의 80%는 데이터와 피처에서 나온다고 할 정도로 중요
- 도메인 지식(업무 이해)이 있을수록 좋은 피처를 만들 수 있음

---
## 피처 엔지니어링이란?

### 핵심 용어 정리

| 용어 | 의미 | 엑셀로 비유하면 |
|------|------|----------------|
| **피처 (Feature)** | 모델에 입력하는 변수 | 엑셀의 **열(컬럼)** |
| **타겟 (Target)** | 예측하고 싶은 값 | 정답에 해당하는 **열** |
| **피처 엔지니어링** | 기존 데이터를 조합해서 새 변수를 만드는 것 | 엑셀에서 수식으로 새 열 만들기 |

### 왜 중요한가요?

> **"좋은 피처가 좋은 모델을 만든다"**

같은 모델이라도 어떤 피처를 넣느냐에 따라 성능이 크게 달라집니다.
현업에서 데이터 분석가의 **핵심 역량**이 바로 이 피처 엔지니어링입니다!

## 5. 인코딩 (범주형 → 숫자)

### 왜 인코딩이 필요한가요?

머신러닝 모델은 **숫자만 이해**할 수 있습니다.  
"서울", "부산" 같은 문자 데이터는 숫자로 바꿔야 모델에 넣을 수 있습니다.

### 인코딩 방법 2가지

| 방법 | 사용 시점 | 예시 |
|------|----------|------|
| **LabelEncoder** | 순서가 있는 범주 | 베이직(0) < 스탠다드(1) < 프리미엄(2) |
| **get_dummies (원핫인코딩)** | 순서가 없는 범주 | 서울, 부산, 대전 → 각각 0/1 컬럼 |

---
## 인코딩이 뭔가요? 왜 필요한가요?

### 핵심 개념

**머신러닝 모델은 숫자만 이해할 수 있습니다!**

"서울", "부산", "남자", "여자" 같은 문자 데이터는 그대로 모델에 넣을 수 없습니다.
그래서 문자를 숫자로 바꾸는 작업이 필요한데, 이것을 **인코딩(Encoding)**이라고 합니다.

### 인코딩 방법 2가지

#### 1. LabelEncoder (레이블 인코딩) — 순서가 있는 범주에 사용

| 원본 값 | 변환 값 | 설명 |
|---------|--------|------|
| 베이직 | 0 | 가장 낮은 등급 |
| 스탠다드 | 1 | 중간 등급 |
| 프리미엄 | 2 | 가장 높은 등급 |

→ 숫자의 크기에 **의미가 있음** (0 < 1 < 2 = 등급 순서)

#### 2. OneHotEncoding (원핫 인코딩) — 순서가 없는 범주에 사용

| 원본 | 서울 | 부산 | 대구 |
|------|------|------|------|
| 서울 | 1 | 0 | 0 |
| 부산 | 0 | 1 | 0 |
| 대구 | 0 | 0 | 1 |

→ 각 카테고리를 별도의 0/1 컬럼으로 만듦

### 잘못 쓰면 어떻게 될까?

만약 순서가 없는 "서울, 부산, 대구"를 레이블 인코딩하면?
- 서울=1, 부산=2, 대구=3
- 모델이 **"대구(3)가 서울(1)보다 3배 크다"**고 잘못 이해합니다!
- 순서가 없는 범주에는 반드시 **원핫 인코딩**을 사용하세요!

## 6. 스케일링

### 왜 스케일링이 필요한가요?

키가 170cm이고 몸무게가 70kg인 사람이 있다고 합시다.  
숫자만 보면 170이 70보다 크니까, 모델이 "키가 더 중요하구나"라고 잘못 판단할 수 있습니다.

**스케일링은 모든 특성을 비슷한 범위로 맞춰주는 작업입니다.**

### 언제 쓰나요?
- 단위가 다른 특성이 있을 때 (cm, kg, 원, m 등)
- 트리 기반 모델(결정트리, 랜덤포레스트)은 안 해도 됨
- 거리 기반 모델(KNN), 회귀, 신경망은 **반드시 해야 함**

### StandardScaler vs MinMaxScaler

| 스케일러 | 방법 | 결과 범위 | 언제 사용 |
|---------|------|----------|----------|
| StandardScaler | 평균=0, 표준편차=1로 변환 | 약 -3 ~ 3 | 일반적인 경우 |
| MinMaxScaler | 최소=0, 최대=1로 변환 | 0 ~ 1 | 값의 범위를 맞추고 싶을 때 |

---
## 스케일링이 뭔가요? 왜 필요한가요?

### 문제 상황

아파트 데이터를 생각해보세요:
- **면적**: 15~80 ㎡ (숫자가 큼)
- **역까지 도보**: 1~15분 (숫자가 작음)

이렇게 단위와 범위가 다르면, 모델이 **숫자가 큰 면적에만 집중**하고 역까지 거리는 무시할 수 있습니다.
스케일링은 **모든 변수를 비슷한 범위로 맞춰주는 작업**입니다.

### 스케일링 방법 비교

| 스케일러 | 변환 방식 | 결과 범위 | 언제 사용? |
|---------|----------|----------|-----------|
